## Matrices

````
A = [[1, 4, 5], 
    [-5, 1, 9],
    [-6, 7, 1]]

print("A =", A) 
print("A[1] =", A[1])      # segunda fila
print("A[1][2] =", A[1][2])   # tercer elemento de la segunda fila
print("A[0][-1] =", A[0][-1])   # últimoelemento de la primera fila
````

# Suma de números en una lista

````
numbers = [6, 5, 3, 8, 4, 2, 5, 4, 11]

# variable de almacenamiento
sum = 0

# interando sobre la lista
for val in numbers:
	sum = sum+val

print("La suma es: ", sum)
````

### Suma de matrices en Python puro:

summatrix.png

En la notación de Einstein:

$$ c_{ij} = a_{ij} + b_{ij} $$

````
# Program to add two matrices using nested loop

X = [[12,7,3],
    [4 ,5,6],
    [7 ,8,9]]

Y = [[5,8,1],
    [6,7,3],
    [4,5,9]]

result = [[0,0,0],
         [0,0,0],
         [0,0,0]]
````

Otro tipo de estructura

```
for indice_1 in range(tamaño):
    for indice_2 in range(tamaño):
      operaciones que desees hacer
```


````
# iterate en las filas
for i in range(len(X)):
   # iterate en las columnas
   for j in range(len(X[0])):
       result[i][j] = X[i][j] + Y[i][j]

for r in result:
   print(r)
````
